### Unipolar/Bipolar
 seems that Unipolar/Bipolar is always BI/UNI OR B/U

```python
unipolar_bipolar = omop_data_info_df[(omop_data_info_df['is_text'] == True) & (pd.notna(omop_data_info_df['Expected range'])&(omop_data_info_df['Expected range'] == 'Unipolar/Bipolar'))][
['Field', 'Expected range', 'Used Values']]
unipolar_bipolar
```

```python
words = unipolar_bipolar['Used Values'].to_list()
unique_words = set([item.strip("[]'").replace("'", "") for item in words])
unique_words
```

```python
replace_dict  = {'B':'BI',
            'U' : 'UNI'}
```

```python
omop_data_info_df.columns
```

```python
cols = unipolar_bipolar['Field'].to_list()
clean_data = data.copy()
for col in cols:
    if col in clean_data.columns:
        clean_data[col] = clean_data[col].replace(replace_dict)



# # Selecciona las filas donde 'Field' está en 'cols'
# mask = omop_data_info_df['Field'].isin(cols)
# # Asigna True en la columna 'is_categorical' para estas filas
# omop_data_info_df.loc[mask, 'is_categorical'] = True
# # Opcional: también puedes asignar otros valores o realizar más operaciones en este filtro
# # Como en tu código, también puedes actualizar otras columnas en el mismo filtro
# omop_data_info_df.loc[mask, 'Expected range'] = 'UNI/BI'
```
